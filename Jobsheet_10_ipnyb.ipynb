# Praktikum 01 - Bubble Sort #

```
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data [j + 1] = data[j + 1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan:", angka)
print("Data setelah Bubble Sort:", bubble_sort(angka))
```

# Praktikum 02 - Selection Sort #

```
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range(i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

# Praktikum 03 - Insertion Sort #

```
def insertion_sort(data):
  n = len(data)
  for i in range(1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

# Praktikum 06 - Sorting Data Barang (Nama & Harga) #

```
def bubble_sort_key(data, key_index=0, reverse=False):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
      else:
        if data[j][key_index] > data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(),))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), reverse=True))
```

# Penugasan 07 - Analisis Kompleksitas #

```
import random

def bubble_sort_count(data):
    n = len(data)
    compare = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            compare += 1
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return compare

def selection_sort_count(data):
    n = len(data)
    compare = 0
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            compare += 1
            if data[j] < data[min_index]:
                min_index = j
        data[i], data[min_index] = data[min_index], data[i]
    return compare

def insertion_sort_count(data):
    n = len(data)
    compare = 0
    for i in range(1, n):
        key = data[i]
        j = i - 1
        while j >= 0:
            compare += 1
            if data[j] > key:
                data[j + 1] = data[j]
                j -= 1
            else:
                break
        data[j + 1] = key

    return compare

sizes = [10, 50, 100]

for size in sizes:
    print(f"\n=== Ukuran Input: {size} ===")

    data = [random.randint(1, 1000) for _ in range(size)]

    print("Bubble Sort      :", bubble_sort_count(data.copy()))
    print("Selection Sort   :", selection_sort_count(data.copy()))
    print("Insertion Sort   :", insertion_sort_count(data.copy()))
```

# Penugasan 08 - Proyek Mini #

```
def insertion_sort(barang, key, ascending=True):
    """
    barang   : list of dict  → [{"nama": ..., "harga": ...}, ...]
    key      : kunci sorting → "nama" atau "harga"
    ascending: True (kecil-ke-besar), False (besar-ke-kecil)
    """
    for i in range(1, len(barang)):
        current = barang[i]
        j = i - 1

        while j >= 0:
            if ascending and barang[j][key] > current[key]:
                barang[j + 1] = barang[j]
            elif not ascending and barang[j][key] < current[key]:
                barang[j + 1] = barang[j]
            else:
                break
            j -= 1

        barang[j + 1] = current

    return barang


barang_list = []
n = int(input("Masukkan jumlah barang: "))

for i in range(n):
    print(f"\nBarang ke-{i+1}")
    nama = input("Nama barang   : ")
    harga = int(input("Harga barang  : "))
    barang_list.append({"nama": nama, "harga": harga})

print("\n--- Pilih Metode Pengurutan ---")
print("1. Berdasarkan Nama")
print("2. Berdasarkan Harga")
pilihan_key = int(input("Pilihan (1/2): "))

if pilihan_key == 1:
    key = "nama"
else:
    key = "harga"

print("\n--- Pilih Arah Pengurutan ---")
print("1. Ascending (A → Z / Murah → Mahal)")
print("2. Descending (Z → A / Mahal → Murah)")
pilihan_arah = int(input("Pilihan (1/2): "))

ascending = pilihan_arah == 1

hasil = insertion_sort(barang_list, key, ascending)

print("\n=== HASIL PENGURUTAN BARANG ===")
for item in hasil:
    print(f"{item['nama']} - Rp{item['harga']}")
```
