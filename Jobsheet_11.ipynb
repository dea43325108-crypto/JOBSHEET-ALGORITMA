# Praktikum 01 - Sequential Search pada Data Angka #

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indeks {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

# Praktikum 02 - Binary Search pada Data Angka #

```python
def binary_search(data, target):
  left, right = 0, len(data) -1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan di index {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

# Praktikum 03 - Sequential Search untuk Data Nama #

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
      return f"Nama {target} ditemukan di indeks {i}"
  return f"Nama {target} tidak ditemukan"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data:", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Eka"))
```

# Praktikum 04 - Studi Kasus Nilai Mahasiswa dengan Binary Search #

```python
def binary_search_nilai(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Nilai {target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data nilai:", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 65))
```

# Praktikum 05 - Perbandingan Waktu Sequential vs Binary Search #

```python
import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return i
  return -1

def binary_search(data, target):
  left, right = 0, len(data) -1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1, 100000))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

# Praktikum 06 - Kombinasi Sorting dan Searching #

```python
import random

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"{target} ditemukan di indeks {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [34, 12, 56, 78, 23, 9, 44, 67, 5, 89]

print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = 56
print(binary_search(urut, target))
```

# Penugasan 03 - Variasi Sequential Search #

```python
def sequential_search(data, target):
    perbandingan = 0

    for i in range(len(data)):
        perbandingan += 1
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i} setelah {perbandingan} kali perbandingan"

    return f"Data {target} tidak ditemukan setelah {perbandingan} kali perbandingan"


angka = [10, 20, 30, 40, 50, 60, 70, 80]
print("Data:", angka)

print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

# Penugasan 04 - Variasi Binary Search #

```python
def binary_search(data, target):
    left, right = 0, len(data) - 1
    langkah = 0

    while left <= right:
        langkah += 1
        mid = (left + right) // 2

        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid} setelah {langkah} langkah"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Data {target} tidak ditemukan setelah {langkah} langkah"


angka = [10, 20, 30, 40, 50, 60, 70, 80]
print("Data:", angka)
print(binary_search(angka, 60))
print(binary_search(angka, 100))
```

# Penugasan 06 - Studi Kasus Barang Toko #

```python
def binary_search_harga(data, target):
    left, right = 0, len(data) - 1
    langkah = 0

    while left <= right:
        langkah += 1
        mid = (left + right) // 2

        if data[mid][1] == target:
            return f"Barang dengan harga {target} ditemukan: {data[mid]} (ditemukan dalam {langkah} langkah)"
        elif data[mid][1] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Barang dengan harga {target} tidak ditemukan (dicari dalam {langkah} langkah)"

barang = [
    ("Buku", 5000),
    ("Pulpen", 3000),
    ("Pensil", 3000),
    ("Penggaris", 5000),
    ("Penghapus", 2000),
    ("Sampul Buku", 7000),
    ("Tempat Pensil", 10000),
]

barang_sorted = sorted(barang, key=lambda x: x[1])

print("=== Daftar Barang Setelah Diurutkan Berdasarkan Harga ===")
for item in barang_sorted:
    print(item)

target = int(input("\nMasukkan harga yang ingin dicari: "))

print(binary_search_harga(barang_sorted, target))
```

# Penugasan 07 - Analisis Kompleksitas #

```python
import time
import random

def sequential_search(data, target):
    for item in data:
        if item == target:
            return True
    return False

def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return True
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return False

def ukur_waktu(data_size):
    data = list(range(data_size))
    target = random.choice(data)

    start = time.time()
    sequential_search(data, target)
    waktu_seq = time.time() - start

    start = time.time()
    binary_search(data, target)
    waktu_bin = time.time() - start

    return waktu_seq, waktu_bin

ukuran_data = [100, 1000, 10000]
hasil = []

print("=== HASIL PENGAMATAN WAKTU ===")
for size in ukuran_data:
    w_seq, w_bin = ukur_waktu(size)
    hasil.append((size, w_seq, w_bin))
    print(f"Data {size}: Sequential = {w_seq:.6f}s | Binary = {w_bin:.6f}s")
```

# Penugasan 08 - Proyek Mini #

```python
def binary_search(data, target):
    left, right = 0, len(data) - 1
    langkah = 0

    while left <= right:
        langkah += 1
        mid = (left + right) // 2

        if data[mid][1] == target:
            return mid, langkah
        elif data[mid][1] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1, langkah

mahasiswa = []
n = int(input("Masukkan jumlah mahasiswa: "))

for i in range(n):
    print(f"\nMahasiswa ke-{i+1}")
    nama = input("Nama  : ")
    nilai = int(input("Nilai : "))
    mahasiswa.append((nama, nilai))

mahasiswa_sorted = sorted(mahasiswa, key=lambda x: x[1])

print("\n=== Data Mahasiswa Setelah Diurutkan (berdasarkan Nilai) ===")
for m in mahasiswa_sorted:
    print(f"{m[0]} - {m[1]}")

target = int(input("\nMasukkan nilai yang ingin dicari: "))

index, langkah = binary_search(mahasiswa_sorted, target)

print("\n=== HASIL PENCARIAN ===")
if index != -1:
    nama, nilai = mahasiswa_sorted[index]
    print(f"Nilai {target} ditemukan pada mahasiswa: {nama}")
    print(f"Ditemukan pada indeks: {index}")
    print(f"Jumlah langkah binary search: {langkah}")
else:
    print(f"Nilai {target} tidak ditemukan.")
    print(f"Binary search dilakukan dalam {langkah} langkah.")
```
