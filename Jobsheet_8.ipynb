# Praktikum 04 - Fungsi dengan *args dan **kwargs #

```python
def jumlahkan(*args):
  return sum(args)

def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")

print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

# Praktikum 07 - Fungsi Decorator #

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print("Setelah fungsi dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")

sapa()
```

# Penugasan 10 - Fungsi Decorator #

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Memulai fungsi...")
    fungsi()
    print("Fungsi selesai dijalankan...")
  return wrapper

@decorator_sapa
def sapa():
  print("Halo, saya Dea Cantika")

sapa()
```

# Penugasan 11 - Studi Kasus Nyata: Program Nilai #

```python
def input_nilai(judul):
    jml = int(input(f"Jumlah {judul}: "))
    return [float(input(f"{judul} ke-{i+1}: ")) for i in range(jml)]

def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

def status_kelulusan(nilai):
    return "Lulus" if nilai >= 60 else "Tidak Lulus"

# Lambda bobot: 40% tugas, 60% ujian
nilai_akhir = lambda rt, ru: 0.4 * rt + 0.6 * ru

# Program utama
tugas = input_nilai("nilai tugas")
ujian = input_nilai("nilai ujian")

rata_tugas = hitung_rata(tugas)
rata_ujian = hitung_rata(ujian)
akhir = nilai_akhir(rata_tugas, rata_ujian)

print(f"\nRata-rata tugas: {rata_tugas:.2f}")
print(f"Rata-rata ujian: {rata_ujian:.2f}")
print(f"Nilai akhir (berbobot): {akhir:.2f}")
print("Status kelulusan:", status_kelulusan(akhir))
```
