# Penugasan 6 - Modul Buatan Sendiri #

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
  return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
  return (f - 32) * 5/9
```

```python
import konversi

print("12 derajat celcius =", konversi.celcius_ke_fahrenheit(12))
print("12 derajat fahrenheit =", konversi.fahrenheit_ke_celcius(12))
```

# Penugasan 8 - Studi Kasus Mini #

```python
%%writefile laporan.py
def tampilkan_ringkasan_kuis(skor_benar, total_soal):
  print(f"Anda menjawab {skor_benar} dari {total_soal} pertanyaan dengan benar.")
```

```python
import random
import laporan

total_soal = 3
skor_benar = 0

print("Selamat datang di Kuis Penjumlahan!")

for i in range(total_soal):
    angka1 = random.randint(1, 10)
    angka2 = random.randint(1, 10)
    jawaban_benar = angka1 + angka2

    print(f"\nSoal ke-{i+1}: Berapakah {angka1} + {angka2}?")
    try:
        jawaban_pengguna = int(input("Jawaban Anda: "))
        if jawaban_pengguna == jawaban_benar:
            skor_benar += 1
            print("Benar! Kerja bagus.")
        else:
            print(f"Salah. Jawaban yang benar adalah {jawaban_benar}.")
    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")
        print(f"Jawaban yang benar adalah {jawaban_benar}.")

print("\nKuis selesai!")
laporan.tampilkan_ringkasan_kuis(skor_benar, total_soal)
```
